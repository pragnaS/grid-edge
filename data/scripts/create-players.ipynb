```python
#generate complex power using random power factor (between .9-.95)
def complex_power(real_power):
    power_factor = random.uniform(.9, .95)
    apparent_power = real_power / power_factor
    reactive_power = sqrt(apparent_power**2 - real_power**2)
    return str(round(real_power,6))+ '+' + str(round(reactive_power,6))+'j'
```

```python
#create player file for 
dict = {6:'june', 7:'july', 8:'aug',9:'sept'}

new_id = 1

for id in ids:
    orig_df = data[data['dataid']==id]
    
    for i in range(6,10):
        df = orig_df[(orig_df['localminute'].dt.month==i)==1]
        
        #average duplicate data for one minute
        df = df.groupby('localminute').mean().reset_index()

        #set index as local minute to use the asfreq function
        df = df.set_index('localminute')
        df = df.asfreq(freq='60S')

        # Fill NaN with mean of non-NaN value before and after it
        df['grid'] = (df['grid'].ffill()+df['grid'].bfill())/2
        df['grid'] = df['grid'].bfill().ffill()

        df = df.reset_index()

        timestamp = f'2018-0{i}-01 00:00:00 CST'

        #create dataframe for player file output
        final_df = pd.DataFrame(index=range(df.shape[0]),columns=range(2))
        final_df.columns = ['time','grid']
        final_df['time'] = '+60s'
        final_df['grid'] = df['grid'] * 1000 * 15 #change to kW, scale by 15
        final_df['grid'] = final_df['grid'].map(complex_power) #convert to complex

        final_df['time'].iloc[0] = timestamp
        print(final_df.head(2))
        final_df.to_csv(f'./summer_players/{dict[i]}/{str(new_id)}.player', header=None, index=None)

        #format csv to have grid + id 
        df = final_df
        df = df.drop('time', 1)
        df['id'] = new_id
        df['month'] = dict[i]
        df.to_csv(f'./input_csv/{dict[i]}/{str(new_id)}.csv', index=None)
        print(df.head(2))

        print(len(df))
    new_id = new_id + 1 
```
