```python
import pandas as pd
import os
import warnings
warnings.filterwarnings('ignore')
```

```python
directory = '.'
for filename in os.listdir(directory):
    if filename.endswith(".csv"):
        df = pd.read_csv(filename)
        df['real'] = df.apply(lambda x: float(str(x['grid']).split("+")[0]),axis=1)
        df.plot(y='real', title=filename)
```
