```python
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
#reading in GridLAB-D DPMU data
dpmu_phaseA_aug = pd.read_csv(r'C:\Users\aniss\grid-edge\data\summer\targets\trimmed_targets\aug\multi_voltageA.csv')
```

```python
time = []
time = dpmu_phaseA_aug['# timestamp']
```

```python
time = pd.Series(time)
time = time.str.split(" ")
```

```python
for index in range(len(time)):
    del time[index][2]
time
```

```python
month = []
hour = []
for index in range(len(time)):
    month.append(datetime.strptime(time[index][0],'%Y-%m-%d'))
    hour.append(datetime.strptime(time[index][1],'%H:%M:%S'))
```

```python
date = []
timestamp = []

for index in range(len(time)):
    for jindex in range(len(time[index])):
        if "-" in time[index][jindex]:
            date.append(time[index][jindex].split("-"))
        if ":" in time[index][jindex]:
            timestamp.append(time[index][jindex].split(":"))
```

```python
date = np.array(date)
date
```

```python
timestamp = np.array(timestamp)
timestamp.shape
```

```python
#separating the individual values into different list
year = []
month = []
day = []
for y in range(len(date)):
    year.append(date[y][0])
    month.append(date[y][1])
    day.append(date[y][2])

hour = []
minute = []
sec = []
for y in range(len(timestamp)):
    hour.append(timestamp[y][0])
    minute.append(timestamp[y][1])
    sec.append(timestamp[y][2])
```

```python
time_df = pd.DataFrame()
```

```python
time_df["year"] = year
time_df["month"] = month
time_df["day"] = day
time_df["hour"] = hour
time_df["min"] = minute
time_df["sec"] = sec
```
