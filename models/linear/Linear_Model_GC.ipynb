```python
#reading DPMU phase A output from gridlabD
dpmu_phaseA = pd.read_csv('multi_voltageA.csv')
dpmu_phaseB = pd.read_csv('multi_voltageB.csv')
dpmu_phaseC = pd.read_csv('multi_voltageC.csv')
```

```python
#within the GridLabD output (DPMU), isolate Node661 
node_661_dpmu_phaseA = pd.DataFrame(dpmu_phaseA, columns=['# timestamp', 'Node661'])
node_661_dpmu_phaseB = pd.DataFrame(dpmu_phaseB, columns=['# timestamp', 'Node661'])
node_661_dpmu_phaseC = pd.DataFrame(dpmu_phaseC, columns=['# timestamp', 'Node661'])
```

```python
# converting dpmu string values to complex dtype
node_661_dpmu_phaseA['Node661'] = node_661_dpmu_phaseA['Node661'].str.replace("i", "j").apply(lambda x: complex(x))
node_661_dpmu_phaseB['Node661'] = node_661_dpmu_phaseB['Node661'].str.replace("i", "j").apply(lambda x: complex(x))
node_661_dpmu_phaseC['Node661'] = node_661_dpmu_phaseC['Node661'].str.replace("i", "j").apply(lambda x: complex(x))
```

```python
# splitting dpmu phase A values into real and imaginary parts
node_661_dpmu_phaseA['timestamp'] = node_661_dpmu_phaseA['# timestamp'].str.replace('CST', '')
node_661_dpmu_phaseA['realA'] = node_661_dpmu_phaseA['Node661'].apply(lambda x : x.real)
node_661_dpmu_phaseA['imagA'] = node_661_dpmu_phaseA['Node661'].apply(lambda x : x.imag)
node_661_dpmu_phaseA = pd.DataFrame(node_661_dpmu_phaseA, columns = ['timestamp', 'realA', 'imagA'])
node_661_dpmu_phaseA
```

```python
# splitting dpmu phase B values into real and imaginary parts
node_661_dpmu_phaseB['timestamp'] = node_661_dpmu_phaseB['# timestamp'].str.replace('CST', '')
node_661_dpmu_phaseB['realB'] = node_661_dpmu_phaseB['Node661'].apply(lambda x : x.real)
node_661_dpmu_phaseB['imagB'] = node_661_dpmu_phaseB['Node661'].apply(lambda x : x.imag)
node_661_dpmu_phaseB = pd.DataFrame(node_661_dpmu_phaseB, columns = ['timestamp', 'realB', 'imagB'])
node_661_dpmu_phaseB
```

```python
# splitting dpmu phase C values into real and imaginary parts
node_661_dpmu_phaseC['timestamp'] = node_661_dpmu_phaseC['# timestamp'].str.replace('CST', '')
node_661_dpmu_phaseC['realC'] = node_661_dpmu_phaseC['Node661'].apply(lambda x : x.real)
node_661_dpmu_phaseC['imagC'] = node_661_dpmu_phaseC['Node661'].apply(lambda x : x.imag)
node_661_dpmu_phaseC = pd.DataFrame(node_661_dpmu_phaseC, columns = ['timestamp', 'realC', 'imagC'])
node_661_dpmu_phaseC
```

```python
#within the PecanStreet data, get all grid values for Node661
node_661_grid = df.loc[df['dataid'] == 661]
#node_661_grid = pd.DataFrame(node661_grid, columns = ['localminute', 'grid'])
node_661_grid['timestamp'] = node_661_grid['localminute'].str.replace('-06', '')
node_661_grid = pd.DataFrame(node_661_grid, columns = ['timestamp', 'grid'])
node_661_grid
```

```python
node_661_player = pd.read_csv('661.csv')
node_661_player.columns
```

```python
#node_661_player['timestamp'] = node_661_player[node_661_player.columns[0]]
node_661_player['complex_player'] = node_661_player['grid']
node_661_player = node_661_player.drop(axis=1, columns=['grid', 'id'])
#node_661_player = pd.DataFrame(node_661_player, columns = ['timestamp', 'complex_player'])
node_661_player
```

```python
node_661_player['complex_player'] = node_661_player['complex_player'].str.replace("i", "j").apply(lambda x: complex(x))
node_661_player['real'] = node_661_player['complex_player'].apply(lambda x : x.real)
node_661_player['imag'] = node_661_player['complex_player'].apply(lambda x : x.imag)
node_661_player = pd.DataFrame(node_661_player, columns = [node_661_player.columns[0], 'real', 'imag']).head(1441).drop(axis=1, columns=['complex_player'])
node_661_player
```

```python
all_phases_dpmu = pd.DataFrame({'realA': node_661_dpmu_phaseA['realA'], 'imagA': node_661_dpmu_phaseA['imagA'], 'realB': node_661_dpmu_phaseB['realB'], 'imagB': node_661_dpmu_phaseB['imagB'], 'realC': node_661_dpmu_phaseC['realC'], 'imagC': node_661_dpmu_phaseC['imagC'] }).head(524709)
all_phases_dpmu
```

```python
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression, Ridge, Lasso, LogisticRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

# create datasets
X = pd.DataFrame(node_661_player, columns=['real', 'imag'])
y = all_phases_dpmu

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
print(X.shape)
print(y.shape)
```

```python
# define model
model = LinearRegression()
#model = Ridge(alpha=0.3)
#model = Lasso(alpha=0.5)
# fit model
model.fit(X_train, y_train)
# make a prediction
y_pred = model.predict(X_test)

print(y_pred)
# summarize prediction
print(mean_absolute_error(y_test, y_pred) * 100)
```
